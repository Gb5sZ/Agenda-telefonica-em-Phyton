```python
agenda = []


#funções que irão receber os dados para as variaveis nome, tel, email, tt & face.


def solicita_nome():
     return(input("Insira seu Nome: "))

def solicita_tel():
     return(input("Insira seu Telefone: "))

def solicita_email():
     return(input("Insira seu e-mail: "))

def solicita_tt():
     return(input("Insira seu @ no Twitter: "))

def solicita_face():
     return(input("Insira seu Facebook: "))


#A função "show_data" é a função que ira listar na tela todos os contatos.


def show_data(nome, tel, email, tt, face):
     print("Nome: %s | Telefone: %s | Email: %s | Twitter: %s | Facebook: %s" % (nome, tel, email, tt, face))

#A função "busca" que sera aproveitada nas funções Apagar, Alterar & Consultar.

def busca(nome):
     mnome = nome.lower()
     for p, e in enumerate(agenda):
         if e[0].lower() == mnome:
               return p
     return None

#A função "criar" ira iniciar o processo de cadastro de um  novo contato para a agenda.

def criar():
     global agenda
     nome = solicita_nome()
     tel = solicita_tel()
     email = solicita_email()
     tt = solicita_tt()
     face = solicita_face()
     agenda.append([nome, tel, email, tt, face])


#A função "exclui" é a função responsavel pela exclusão de um contato existente na agenda,
#  a busca para que seja realizada a exclusão sera realizada com base no nome cadastrado.


def exclui():
     global agenda
     nome = solicita_nome()
     p = busca(nome)
     if p != None:
         del agenda[p]
     else:
         print("Nome não encontrado.")


#A função "altera" é responsavel pela alteração dos dados de um contato existente, 
# e funciona da mesma maneira realizando a busca do contato que deseja ser alterado por meio do nome registrado.


def altera():
     p = busca(solicita_nome())
     if p != None:
         nome = agenda[p][0]
         tel = agenda[p][1]
         email = agenda [p][2]
         tt = agenda [p][3]
         face = agenda [p][4]
         print("Encontrado:")
         show_data(nome, tel, email, tt, face)
         nome = solicita_nome()
         tel = solicita_tel()
         email = solicita_email()
         tt = solicita_tt()
         face = solicita_face()
         agenda[p] = [nome, tel, email, tt, face]
     else:
         print("Nome não encontrado.")


#A função "lista" é a função responsavel por listar todos os contatos existentes na agenda.


def lista():
     print("\n=============Agenda=====================================")
     for e in agenda:
         show_data(e[0], e[1], e[2], e[3], e[4])
     print("\n=============Agenda=====================================")


#A função "consulta" é responsavel por consultar ou seja ela ira listar um contato desejado,
#caso ele exista com base na busca pelo nome registrado.


def consulta():
     p = busca(solicita_nome())
     if p != None:
         nome = agenda[p][0]
         tel = agenda[p][1]
         email = agenda [p][2]
         tt = agenda [p][3]
         face = agenda [p][4]
         print("Encontrado:")
         show_data(nome, tel, email, tt, face)         
     else:
         print("Nome não encontrado.")


#A função "verifica_escolha" é responsavel por garantir que a agenda funcione apenas se o
#usuario escolher o numero que corresponde a uma respectiva função, os numeros de 0 a 5 caso contrario emitira um alerta.


def verifica_escolha(pergunta, inicio, fim):
     while True:
         try:
               valor = int(input(pergunta))
               if inicio <= valor <= fim:
                   return(valor)
         except ValueError:
               print("Valor inválido, favor digitar entre %d e %d" % (inicio, fim))
                
                
#A função "menu" é a aonde estão todas as variaveis de escolha
#e sua interface com as informações necessarias para identificação


def menu():
     print("""
______________________________________
|   1 - Cria um Novo Registro        |
|   2 - Alterar Registro existente   |
|   3 - Excluir Registro existente   |
|   4 - Listar Registros existentes  |
|   5 - Pesquisa Registro especifico |
|                                    |
|   0 - Sai                          |
L____________________________________⅃
""")
     return verifica_escolha("Escolha uma opção: ",0,5)

while True:
     opcao = menu()
     if opcao == 0:
         break
     elif opcao == 1:
         criar()
     elif opcao == 2:
         altera()
     elif opcao == 3:
         exclui()
     elif opcao == 4:
         lista()
     elif opcao == 5:
         consulta()
```
